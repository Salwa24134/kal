## BAB 5 Tranformasi Matriks
### 5.1 Tranformasi Matrik
Tranformasi matrik adalah bentuk khusus dari tranformasi linear, yang mengubah satu vektor menjadi vektor lain melalui perkalian matrik. Visualisasi ini sangat berguna terutama dalam bidang seperti grafika komputer. Definisi aljabar memungkinkan kita menggunakan matriks untuk mendefinisikan fungsi yang mengubah satu vektor menjadi vektor lain, seperti fungsi dalam matematika yang mengubah satu angka menjadi angka lain, kemudian bisa memvisualisasikan matriks dan perkalian matriks berdasarkan pengaruhnya terhadap vektor.Diberikan sebuah matriks A berukuran m x n, kita bisa mendefinisikan fungsi yang menerima vektor kolom x berukuran n x 1 sebagai input, dan menghasilkan vektor kolom Ax berukuran m x 1 sebagai output, sesuai hubungan: $\vec{y} = T(\vec{x}) = A\vec{x}$.
Fungsi semacam ini dikenal sebagai transformasi matrik, dan merupakan contoh dari kelas fungsi yang lebih umum antara ruang vektor, yang disebut transformasi linear.


#### 5.1.1 Perkalian Matriks dengan Vektor
Perkalian matriks dengan vektor dapat divisualisasikan secara geometris terutama dalam ruang dua dimensi (R²). Dalam contoh yang diberikan, beberapa vektor dikalikan dengan sebuah matriks 2×2, menghasilkan vektor baru dengan arah dan panjang yang berbeda. Ada vektor yang tetap searah setelah transformasi, sementara lainnya berubah arah. Hal ini menunjukkan bahwa matriks bertindak sebagai alat transformasi yang dapat memutar, memperbesar, memperkecil, atau bahkan menghilangkan vektor. Tidak semua matriks memengaruhi vektor dengan cara yang sama—beberapa memperpanjang semua vektor, beberapa memperpendek, dan ada juga yang hanya mengubah arah vektor tertentu. Untuk memahami bagaimana matriks memengaruhi vektor, digunakan pendekatan dengan menggabungkan operasi perkalian matriks dan penjumlahan vektor, yang juga divisualisasikan melalui grafik.

#### 5.1.2 : Transformasi pada Bidang Kartesius


Transformasi matriks tidak hanya memengaruhi satu vektor, tetapi juga dapat diterapkan ke seluruh bidang Kartesius. Karena sifat distributif dari perkalian matriks terhadap penjumlahan vektor, bidang ini akan mengalami perubahan secara konsisten—garis lurus tetap menjadi garis lurus, dan kurva tetap menjadi kurva meskipun bentuknya bisa berubah. Salah satu cara efektif untuk memahami dampak transformasi ini adalah dengan mengamati bagaimana matriks mengubah bentuk dari unit square (persegi satuan). Ketika keempat titik sudut dari persegi tersebut dikalikan dengan matriks, hasilnya menunjukkan bagaimana seluruh bidang akan berubah. Proses ini memberikan pemahaman visual yang kuat mengenai bagaimana transformasi matriks memengaruhi posisi dan bentuk dari kumpulan vektor.

#### 5.1.3  Kombinasi Penjumlahan dan Perkalian Matriks

Interaksi antara penjumlahan vektor dan perkalian matriks menunjukkan bahwa sifat distributif tetap berlaku. Dalam contoh yang diberikan, penjumlahan dua vektor dilakukan terlebih dahulu, kemudian hasilnya dikalikan dengan matriks. Ternyata hasilnya sama dengan jika masing-masing vektor dikalikan terlebih dahulu, lalu dijumlahkan. Secara visual, hal ini diperjelas dengan grafik vektor dan bentuk paralelogram yang terbentuk. Kesimpulannya, transformasi matriks mempertahankan struktur aljabar dari vektor, khususnya sifat distributif terhadap penjumlahan. Pemahaman ini menjadi dasar untuk mengenali bahwa transformasi matriks adalah bentuk khusus dari transformasi linear.
